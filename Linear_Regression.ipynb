```python
x = df.drop(columns="package")
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
lr.fit(x_train,y_train)

```

```python
x_test.iloc[0]
```

```python
lr.predict(x_test.iloc[[0]])
```

```python
plt.scatter(df["cgpa"],df["package"],color="red")
plt.plot(x_train,lr.predict(x_train),color="black")
plt.xlabel("CGPA")
plt.ylabel("PACKAGE")
plt.show()
```

```python
m = lr.coef_
b = lr.intercept_
```

```python
x = float(input("CGPA"))
y = m*x + b
print("your package will be around :" , y)
```
